## Підготовка

```python
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import utils
from tensorflow.keras.preprocessing import image
from google.colab import files
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline 
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
classes = ['tshirt', 'pants', 'sweater', 'dress', 'coat', 'shoe', 'shirt', 'sneaker', 'bag', 'boot']
```

Змінюємо формат даних, нормалізовуємо дані і перетворюємо в формат one hot encoding

```python
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)

x_train = x_train / 255 
x_test = x_test / 255 

y_train = utils.to_categorical(y_train, 10)
y_test = utils.to_categorical(y_test, 10)
```

```python
def create_neural(input_layer, hidden_layers, epochs=100, batch_size=200):
  model = Sequential()
  model.add(Dense(input_layer, input_dim=784, activation='relu'))
  for layer in hidden_layers:
    model.add(Dense(layer, activation="relu"))
  model.add(Dense(10, activation="softmax"))
  model.compile(loss="categorical_crossentropy", optimizer="SGD", metrics=["accuracy"])

  

  history = model.fit(
      x_train, y_train, 
      batch_size=batch_size, # 200
      epochs=epochs, # 100
      validation_split=0.2,
      verbose=1
  )

  return model
```

```python
def check_models(params, epochs=100, batch_size=200):
  results = []
  for param in params:
    in_layer = param[0]
    hidden = param[1]

    model = create_neural(in_layer, hidden, epochs=epochs, batch_size=batch_size)
    score = model.evaluate(x_test, y_test, verbose=1)

    results.append(score[1])
  return results
```

```python
def print_results(params, results):
  for i in range(len(results)):
    param = params[i]
    result = results[i]
    print("Input layer size:", param[0], end=', ')
    if(len(param[1]) == 0): 
      print("no hidden layers", end='')
    else:
      print("with hidden layers: [", end='')
      for i in range(len(param[1])):
        if(i > 0):
          print(", ", end='')
        print(param[1][i], end='')
      print("]", end=',')  
    print("got result:", f'{result*100:.2f}%')

```
